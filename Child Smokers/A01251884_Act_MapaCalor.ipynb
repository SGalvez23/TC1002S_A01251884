```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd  
import seaborn as sns
```

```python
cs = pd.read_csv('Child smokers.csv', na_values = ':')
cs
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.distplot(cs['Age (years)'], bins = 48)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.distplot(cs['Height (cm)'], bins = 48)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.distplot(cs['FEV (litres)'], bins = 48)
plt.show()
```

```python
correlation_matrix = cs.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
cs = cs.dropna()
```

```python
fig = plt.figure(figsize = (15, 10))
plt.boxplot(cs['Age (years)'], notch=True)
plt.title('Edad')
plt.show()
```

```python
fig = plt.figure(figsize = (15, 10))
plt.boxplot(cs['Height (cm)'], notch=True)
plt.title('Altura')
plt.show()
```

```python
fig = plt.figure(figsize = (15, 10))
plt.boxplot(cs['FEV (litres)'], notch=True)
plt.title('FEV')
plt.show()
```
